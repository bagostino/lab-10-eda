```python
# Manual calculation: population variance and standard deviation
# Note: Pandas uses sample standard deviation by default, so results may differ slightly.

```

```python
mean = housing['Sale_Price'].mean()
squared_diffs = (housing['Sale_Price'] - mean) ** 2
variance = squared_diffs.sum() / len(housing['Sale_Price'])
population_std = variance ** 0.5

print(variance)
print(population_std)
```

```python
housing['Sale_Price'].var()

```

```python
housing['Sale_Price'].std()
```

Is there a wide spread in sale prices? Std deviation is almost $80k-- there is a wide spread in sale prices.
